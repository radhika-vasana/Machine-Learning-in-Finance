```python
import pandas as pd

data = pd.DataFrame({
    "relationship": ["Husband", "Wife", "Own-child", "Not-in-family", "Husband", "Wife"],
    "hours-per-week": [40, 50, 60, 20, 55, 45]
})

n = 10

def reduce_hours(x):
    return x - n

data["reduced_hours"] = data["hours-per-week"].apply(reduce_hours)

grouped_original = data.groupby(["relationship", "hours-per-week"]).size().reset_index(name="count")

grouped_reduced = data.groupby(["relationship", "reduced_hours"]).size().reset_index(name="count")

print("Original Grouping (relationship + hours-per-week):")
print(grouped_original, "\n")

print("Reduced Grouping (relationship + reduced_hours):")
print(grouped_reduced)
```
